```python
#null hypothesis is that there is no increase in the sales after implementing digital marketing, and since the sample size is more than 20, we can adopt Z-Test
```

```python
zstat1,p_val1=ztest(data["Sales_before_digital_add(in $)"],data['Sales_After_digital_add(in $)'],value=0)
p_val1
```

```python
if p_val1<.05:
    print('null hypothes is rejected')
else:
    print('null hypothesis is accepted ')
```
